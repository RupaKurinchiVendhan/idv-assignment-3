```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV (no header) and parse
df = pd.read_csv("transformed_COVID_monthly_cases_2020.csv", header=None)

# First row: months (YYYY-MM), second row: case values
months = df.iloc[0].tolist()
cases = df.iloc[1].astype(float).values
month_labels = pd.to_datetime(months, format='%Y-%m').strftime('%b\n%Y')

# Parameters
N = len(cases)
angles = np.linspace(0, 2 * np.pi, N, endpoint=False)
max_val = max(cases)
buffer = max_val * 0.2
total_height = max_val + buffer

# Create polar plot
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={'polar': True})
ax.set_theta_zero_location('N')
ax.set_theta_direction(-1)

# Plot bars with buffer at center
bars = ax.bar(angles, cases, width=2*np.pi/N, align='edge',
              bottom=buffer,
              color='lightcoral', edgecolor='darkred', linewidth=1.5)

# Add month labels (pushed further out)
for angle, label in zip(angles, month_labels):
    rotation = np.degrees(angle)
    alignment = 'left' if np.pi/2 < angle < 3*np.pi/2 else 'right'
    ax.text(angle, total_height * 1.15, label,
            ha=alignment, va='center',
            rotation=rotation, rotation_mode='anchor', fontsize=9)

# Add concentric ring gridlines and labels (value key)
ring_values = np.linspace(buffer, max_val + buffer, 4)[1:]  # skip inner buffer
ax.set_yticks(ring_values)
ax.set_yticklabels([f"{int(v - buffer):,}" for v in ring_values], fontsize=9)

# Optional: style gridlines
ax.yaxis.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.5)

# Hide x-axis ticks and frame
ax.set_xticks([])
ax.spines['polar'].set_visible(False)

# Title
plt.title("Monthly New Confirmed COVID-19 Cases", fontsize=16, weight='bold', y=1.1)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# List your CSVs here
csv_files = ["transformed_COVID_monthly_cases_2020.csv", "transformed_COVID_monthly_cases_2021.csv", "transformed_COVID_monthly_cases_2022.csv"]
titles = ["Region A", "Region B", "Region C"]  # Customize as needed

# First, find the global maximum new_confirmed_cases across all CSVs
all_cases = []
for file in csv_files:
    df = pd.read_csv(file, header=None)
    cases = df.iloc[1].astype(float).values
    all_cases.append(cases)

global_max = max([max(c) for c in all_cases])
buffer = global_max * 0.2
total_height = global_max + buffer

# Set up subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 8), subplot_kw={'polar': True})
fig.subplots_adjust(wspace=0.4)

def plot_radial_chart(ax, months, cases, title):
    month_labels = pd.to_datetime(months, format='%Y-%m').strftime('%b\n%Y')
    N = len(cases)
    angles = np.linspace(0, 2 * np.pi, N, endpoint=False)

    # Orientation
    ax.set_theta_zero_location('N')
    ax.set_theta_direction(-1)

    # Bars
    ax.bar(angles, cases, width=2*np.pi/N, align='edge',
           bottom=buffer,
           color='lightcoral', edgecolor='darkred', linewidth=1.5)

    # Month labels
    for angle, label in zip(angles, month_labels):
        rotation = np.degrees(angle)
        alignment = 'left' if np.pi/2 < angle < 3*np.pi/2 else 'right'
        ax.text(angle, total_height * 1.15, label,
                ha=alignment, va='center',
                rotation=rotation, rotation_mode='anchor', fontsize=8)

    # Concentric rings
    ring_values = np.linspace(buffer, global_max + buffer, 4)[1:]
    ax.set_yticks(ring_values)
    ax.set_yticklabels([f"{int(v - buffer):,}" for v in ring_values], fontsize=8)
    ax.set_ylim(0, total_height * 1.3)
    ax.yaxis.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.5)

    ax.set_xticks([])
    ax.spines['polar'].set_visible(False)
    ax.set_title(title, fontsize=14, weight='bold', y=1.1)

# Loop through axes and plot data
for ax, file, title, case_vals in zip(axes, csv_files, titles, all_cases):
    df = pd.read_csv(file, header=None)
    months = df.iloc[0].tolist()
    plot_radial_chart(ax, months, case_vals, title)

plt.suptitle("Monthly New Confirmed COVID-19 Cases Across Regions", fontsize=18, weight='bold', y=1.05)
plt.tight_layout()
plt.show()

# Set up subplots (1 row, 3 columns), each with polar coordinates
fig, axes = plt.subplots(1, 3, figsize=(18, 8), subplot_kw={'polar': True})
fig.subplots_adjust(wspace=0.4)  # spacing between plots

# Function to process and plot a single radial chart
def plot_radial_chart(ax, csv_file, title):
    df = pd.read_csv(csv_file, header=None)
    months = df.iloc[0].tolist()
    cases = df.iloc[1].astype(float).values
    month_labels = pd.to_datetime(months, format='%Y-%m').strftime('%b\n%Y')

    N = len(cases)
    angles = np.linspace(0, 2 * np.pi, N, endpoint=False)
    max_val = max(cases)
    buffer = max_val * 0.2
    total_height = max_val + buffer

    # Set orientation
    ax.set_theta_zero_location('N')
    ax.set_theta_direction(-1)

    # Draw bars
    ax.bar(angles, cases, width=2*np.pi/N, align='edge',
           bottom=buffer,
           color='lightcoral', edgecolor='darkred', linewidth=1.5)

    # Add month labels
    for angle, label in zip(angles, month_labels):
        rotation = np.degrees(angle)
        alignment = 'left' if np.pi/2 < angle < 3*np.pi/2 else 'right'
        ax.text(angle, total_height * 1.15, label,
                ha=alignment, va='center',
                rotation=rotation, rotation_mode='anchor', fontsize=8)

    # Add concentric ring labels
    ring_values = np.linspace(buffer, max_val + buffer, 4)[1:]
    ax.set_yticks(ring_values)
    ax.set_yticklabels([f"{int(v - buffer):,}" for v in ring_values], fontsize=8)
    ax.yaxis.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.5)

    ax.set_xticks([])
    ax.spines['polar'].set_visible(False)
    ax.set_title(title, fontsize=14, weight='bold', y=1.1)

# Loop over files and axes
for ax, file, title in zip(axes, csv_files, titles):
    plot_radial_chart(ax, file, title)

plt.suptitle("Monthly New Confirmed COVID-19 Cases Across Regions", fontsize=18, weight='bold', y=1.05)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

def transform_covid_data_weekly(input_file):
    # Read the original CSV
    df = pd.read_csv(input_file)

    # Convert 'date' column to datetime format
    df["date"] = pd.to_datetime(df["date"])

    # Extract year-week as string (ISO calendar week)
    df["year_week"] = df["date"].dt.strftime("%G-W%V")  # ISO week format

    # Group by year-week and average new cases
    weekly_averages = df.groupby("year_week")[["new_confirmed", "new_recovered", "new_deceased"]].mean()

    # Extract year from the week identifier
    weekly_averages["year"] = weekly_averages.index.str[:4]

    # Split into separate years and save
    for year in ["2020", "2021", "2022"]:
        year_df = weekly_averages[weekly_averages["year"] == year].drop(columns="year")
        transformed_df = year_df.T  # Transpose: categories as rows, weeks as columns

        output_file = f"transformed_COVID_weekly_{year}.csv"
        transformed_df.to_csv(output_file)
        print(f"Saved weekly average file for {year}: {output_file}")

# Example usage:
input_csv = "COVID_US_cases.csv"
transform_covid_data_weekly(input_csv)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# File paths for the weekly transformed data
csv_files = [
    "transformed_COVID_weekly_2020.csv",
    "transformed_COVID_weekly_2021.csv",
    "transformed_COVID_weekly_2022.csv"
]
titles = ["2020", "2021", "2022"]

# Load all data and compute global max for uniform scaling
all_cases = []
all_weeks = []

for file in csv_files:
    df = pd.read_csv(file, index_col=0)
    weekly_cases = df.loc["new_confirmed"].values
    weeks = df.columns.tolist()
    all_cases.append(weekly_cases)
    all_weeks.append(weeks)

global_max = max([max(c) for c in all_cases])
buffer = global_max * 0.2
total_height = global_max + buffer

# Setup subplots (3 side-by-side radial plots)
fig, axes = plt.subplots(1, 3, figsize=(18, 8), subplot_kw={'polar': True})
fig.subplots_adjust(wspace=0.4)

# Plot function
def plot_radial_weekly(ax, weeks, values, title):
    N = len(values)
    angles = np.linspace(0, 2 * np.pi, N, endpoint=False)

    # Polar settings
    ax.set_theta_zero_location('N')
    ax.set_theta_direction(-1)

    # Bars
    ax.bar(angles, values, width=2*np.pi/N, align='edge',
           bottom=buffer,
           color='lightcoral', edgecolor='darkred', linewidth=1.5)

    # Add week labels every ~8 weeks to avoid clutter
    label_step = max(1, N // 10)
    for i in range(0, N, label_step):
        angle = angles[i]
        week_label = weeks[i]
        rotation = np.degrees(angle)
        alignment = 'left' if np.pi/2 < angle < 3*np.pi/2 else 'right'
        ax.text(angle, total_height * 1.12, week_label,
                ha=alignment, va='center',
                rotation=rotation, rotation_mode='anchor', fontsize=7)

    # Ring gridlines and labels
    ring_values = np.linspace(buffer, global_max + buffer, 4)[1:]
    ax.set_yticks(ring_values)
    ax.set_yticklabels([f"{int(v - buffer):,}" for v in ring_values], fontsize=8)
    ax.set_ylim(0, total_height * 1.3)
    ax.yaxis.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.5)

    ax.set_xticks([])
    ax.spines['polar'].set_visible(False)
    ax.set_title(title, fontsize=14, weight='bold', y=1.1)

# Plot each subplot
for ax, weeks, values, title in zip(axes, all_weeks, all_cases, titles):
    plot_radial_weekly(ax, weeks, values, title)

plt.suptitle("Weekly Average New Confirmed COVID-19 Cases (2020–2022)", fontsize=18, weight='bold', y=1.05)
plt.tight_layout()
plt.show()
```
